# Benchmark

```python
oos_date = "2020-01-01"

files = {
    "buy_and_hold_positive_ml": "csv\\buy_and_hold_positive_ML.csv",
    "ml": "csv\\buy_and_hold_ML.csv",
    "multi_horizon_trend_confirmed_ML": "csv\\multi_horizon_trend_confirmed_ML.csv"
}

benchmark_file = "csv\\buy_and_hold_snp.csv"
beta_file = benchmark_file

benchmark_metrics = Metrics(
    benchmark_file,
    oos_date=oos_date,
    risk_free_rate=0.00,
    remove_initial_stagnent=True,
    beta_series_file=beta_file,
)

benchmark_metrics.run()
benchmark_metrics.display_table()
benchmark_metrics.plot_nav(log_scale=False)

strategy_metrics_list = []

for name, filename in files.items():
    print(f"\n===== Running strategy: {name} =====")

    m = Metrics(
        filename,
        oos_date=oos_date,
        risk_free_rate=0.00,
        beta_series_file=beta_file,
        remove_initial_stagnent=True,
    )

    m.run()
    m.display_table()
    m.plot_nav(log_scale=False)

    strategy_metrics_list.append(m)
    
outer_joined_df = plot_multiple_series_v2(
    *strategy_metrics_list,
    benchmark_metrics
)
```

## Strategy Comparison Plots and Tables ##

### Preprocessing for merged dataframe ###

```python
from objs.metrics import preprocess_multi_series

processed_series, strategies = preprocess_multi_series(outer_joined_df)
```
